```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL page we will scraping (see image above)
url = "https://www.boxofficemojo.com/chart/ww_top_lifetime_gross"
# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html,'lxml')
```

```python
# use findALL() to get the column headers
soup.findAll('tr', limit=2)
# use getText()to extract the text we need into a list
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
# exclude the first column as we will not need the ranking order from Boxoffice Reference for the analysis
headers = headers[0:]
headers
```

```python
# avoid the first header row
rows = soup.findAll('tr')[1:]
movies = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
movie = pd.DataFrame(movies, columns = headers)
movie.head(10)
```

```python
movie.to_csv("boxoffice_imdb_movie.csv", index=False)
```
