```python
import json
import requests
from bs4 import BeautifulSoup as bts
import bs4
import numpy as np
import pandas as pd
import requests
```

```python
def parseUrl(url):
    result=requests.get(url,headers={'User-Agent': "Mozilla/5.0"})
    soup=bts(result.text,'html.parser')
    return soup
```

```python
#Top 1000 Highest-Grossing Movies of All Time
TOTAL_LINK=[]
def find_total_link():
    for totallink in range(1,11):
          TOTAL_LINK.append("https://www.imdb.com/list/ls098063263/?sort=list_order,asc&st_dt=&mode=detail&page="+str(totallink))
    return TOTAL_LINK
find_total_link()
```

```python
len(TOTAL_LINK)
```

```python
ALL_PRODUCT_URL=[]
def all_product():
    for i in TOTAL_LINK:
        html=parseUrl(i)

        
        for link in  html.findAll('h3',class_='lister-item-header'):
            ALL_PRODUCT_URL.append("https://www.imdb.com/"+link.a["href"])
            
    return ALL_PRODUCT_URL
all_product()
```

```python
len(ALL_PRODUCT_URL)
```

```python
RESULT=[]
for details in ALL_PRODUCT_URL:
    html=parseUrl(details)
    
    ############## Movie Name ##############
    name = html.find('div', {'class': 'sc-94726ce4-1 iNShGo'}).h1.text
    
    ############## Release Year ##############
    Year = html.find('span',{'class':'sc-8c396aa2-2 itZqyK'})
    if Year is None:
            year=np.NaN
    else:
            year=Year.text
    
    
    ############## Runtime ##############
    #runtime=html.find('ul',{'class':'ipc-metadata-list ipc-metadata-list--dividers-none ipc-metadata-list--compact ipc-metadata-list--base'}).find('li',{'class':'ipc-metadata-list__item'}).text.replace('Runtime','').strip()
    run_ul= html.find('ul', { 'class': 'ipc-metadata-list ipc-metadata-list--dividers-none ipc-metadata-list--compact ipc-metadata-list--base'})
    if run_ul is None:
        print("None")
        continue

    run_li= run_ul.findAll('li',{'class':'ipc-metadata-list__item'})
    if len(run_li)<2:
        print("None")
        continue

    runtime = run_li[0].find("div").text
    
    ############## Movie Director ##############
    director = html.find('a', {'class': 'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link'}).text

    ############## Movie Genre ##############
    Genres=html.find('div',{'data-testid':'genres'})
    if Genres is None:
        continue
    sp=Genres.findAll("span")
    if sp is None:
        sp = np.NaN
    elif len(sp)==3:
        genre1 = sp[0].text
        genre2 = sp[1].text
        genre3 = sp[2].text
    elif len(sp)==2:
        genre1 = sp[0].text
        genre2 = sp[1].text
        genre3 = "-"
    else:
        genre1 = sp[0].text
        genre2 = "-"
        genre3 = "-"
    
    ############## Rating ##############
    Rating=html.find('span',{'class':'sc-7ab21ed2-1 jGRxWM'})
    if Rating is None:
            rating=np.NaN
    else:
            rating=Rating.text

    ############## User and Critic Reviews ##############
    #user_reviews=html.find('ul',{'class':'ipc-inline-list sc-124be030-0 ddUaJu baseAlt'}).findAll('li',{'class':'ipc-inline-list__item sc-124be030-1 ghlYSH'})[0].find('span').text.replace('User reviews','').strip()
    #critic_reviews=html.find('ul',{'class':'ipc-inline-list sc-124be030-0 ddUaJu baseAlt'}).findAll('li',{'class':'ipc-inline-list__item sc-124be030-1 ghlYSH'})[1].find('span').text.replace('Critic reviews','').strip()
    
    ul_ = html.find('ul', { 'class': 'ipc-inline-list sc-124be030-0 ddUaJu baseAlt'})
    if ul_ is None:
        print("None")
        continue

    li_ = ul_.findAll('li',{'class':'ipc-inline-list__item sc-124be030-1 ghlYSH'})
    if len(li_)<2:
        print("None")
        continue

    user_reviews = li_[0].find("span").span.text
    critic_reviews = li_[1].find("span").span.text
    
    
    ############## METASCORE ##############
    Metascore=html.find('span',{'class':'score-meta'})
    if Metascore is None:
        metascore=np.NaN
    else:
        metascore=Metascore.text
    
    
    ############## BUDGET ##############    
    Budget=html.find('li',{'data-testid':'title-boxoffice-budget'})
    if Budget is None:
            budget = np.NaN
    else:
            budget = Budget.li.text
    
     ############## GROSS ############## 
    Gross = html.find('li', {'class': 'ipc-metadata-list__item sc-3c7ce701-2 eYXppQ', 
                              'data-testid': 'title-boxoffice-cumulativeworldwidegross'})
    if Gross is None:
            gross=np.NaN
    else:
            gross = Gross.find("ul").findAll("li")[0].text

    #print(gross)
    #print(certificates)
    
    
    
    
    RESULT.append([name,year,runtime,director,genre1,genre2,genre3,rating,user_reviews,critic_reviews,metascore,budget,gross])
columns = ["movie_name","release_year","runtime","director","genre1","genre2","genre3","rating","user_reviews","critic_reviews","metascore","budget","gross"]
df = pd.DataFrame.from_records(RESULT, columns=columns)

        
```
